```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data/house_price_Dống-Da_Hà-Nội_subdata.csv", encoding = "ISO-8859-1")
df
```

```python
df.info()
```

```python
df1 = df[df['area'].notna()]
df1
```

```python
df1 = df1[df['price'].notna()]
df1
```

```python
df1.info()
```

```python
df1=df1.drop_duplicates()
df1
```

## Biểu đồ giá nhà với diện tích

```python
sns.lmplot(x='area',y='price',data = df1)
```

### N/x: Diện tích càng lớn, giá nhà càng cao

## Biểu đồ giá nhà với số lượng phong ngủ

```python
sns.lmplot(x='bedroom',y='price',data = df1)
```

### N/x: nhà có giá càng cao thì thì số phòng ngủ càng có xu hướng tăng

### Biểu đồ giá nhà theo các hướng

```python
sns.violinplot(x= 'house_direction',y= 'price', data= df1)
```

### Giá nhà cao nhất ở hướng Tây - Bắc, thấp nhất ở hướng Bắc. Hướng Đông Bắc có giá trung bình cao nhất rồi đến ĐN, TB, 3 hướng còn lại rất thấp

## Biểu đồ tần số số nhà theo hướng nhà

```python
sns.countplot(x='house_direction',data= df1)
```

## Hướng nhà ĐN được rao nhiều nhất hơn hẳn các hướng khác, hướng bắc rao ít nhất

## Vẽ Box plot phân bố giá nhà theo các hướng

```python
sns.boxplot(x='price',y='house_direction', data= df1)
```

## hướng nhà ĐN có phân bố ngoại lai với giá cao vượt trội, các hướng còn lại giá khá ổn định
